## Matplotlib example workflow 

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # width * hight of the plot

# 3. Plot data - add data to axes
ax.plot(x,y)

# 4. Customise plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel="y-axis")

# 5. Save & show ( you can save the whole figure)
fig.savefig("exported-plots/sample-plot.png")
```

## Common types of Visualisation Plot 
### Making figures with NumPy arrays

### 1. Line Plot

```python
import numpy as np
```

```python
#create some data
# Return evenly spaced numbers over a specified interval.
x = np.linspace(1,10,100)
x[:10]
```

```python
#plot the data and create a line plot

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,x**2); 

ax.set(title="sample-plot2",xlabel="x-axis",ylabel='y-axis')
```

### 2. Scatter Plot

```python
# scatter plot

fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))

ax.set(title='Scatter Plot')
```

```python
# another scatter plot

fig,ax = plt.subplots()
ax.scatter(x,np.sin(x)) # sin wave
ax.plot();
```

### 3. Bar Plot

```python
# Using dictionary to create a plot

nut_butter_prices = {"Almond Butter":10,
                    "Peanut Butter":8,
                    "Cashew Butter":12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())

ax.set(title="Dan's Nut Butter Prices",ylabel="Price ($)");
```

```python
# horizontal bar graph

fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Butter Prices",xlabel="Prices ($)");
```

### 4. Histogram

```python
# make some data

# Return a sample (or samples) from the "standard normal" distribution.
x = np.random.randn(1000)
fig,ax = plt.subplots(1,2,figsize=(20,10))

ax[0].plot(x)
ax[1].hist(x)

ax[0].set(title="Normal Distribution",xlabel="Value",ylabel="Frequence");
ax[1].set(title="Normal Distribution",xlabel="Value",ylabel="Frequence");
```

### Two options for Subplot
#### Multiple plots at the same time



```python
# subplot option 1 - tuple way

fig, ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each diffrent axis

ax0.plot(x,x/2);
ax1.scatter(np.random.random(10),np.random.random(10))
ax2.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax3.hist(np.random.randn(1000))
```

```python
# subplot option 2 - list index

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each different way 
ax[0,0].plot(x,np.sin(x))
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? pyplot vs matplotlib OO method? 

* When plotting something quickly use Pyplot method
* When plotting something more advanced use the OO method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# pyplot method

# Plot over 50 year people a scatterp plot with age and cholesterols and colour with 
# the target coloum  - one colour for folks having a heart disease

over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target'); # c is for colour

# by default gives a gray scale
```

```python
# OO method + Pyplot
# Plot over 50 year people a scatterp plot with age and cholesterols and colour with 
# the target coloum  - one colour for folks having a heart disease

fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

# setting the range of the x axis
# ax.set_xlim([45,95])
```

```python
# OO method - only
# Plot over 50 year people a scatterp plot with age and cholesterols and colour with 
# the target coloum  - one colour for folks having a heart disease

## oo from scratch
fig,ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50['target']);
#customise the plot
ax.set(title='Heart Disease add Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol')

#adding legends - unpack all legends present in the scatter plot
ax.legend(*scatter.legend_elements(),title="Target")


# adding horizontal line to indicate the average Cholesterol level

ax.axhline(over_50['chol'].mean(),
          linestyle="--",color='r');
```

```python
# subplot of chol, age, thalach

# plot
fig, (ax,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

# add data
scatter = ax.scatter(
            x=over_50['age'],
            y=over_50['chol'],
            c= over_50['target'])

# customise the graph
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol')

#adding legend
ax.legend(*scatter.legend_elements(),title='Target')

# adding the line 
ax.axhline(over_50['chol'].mean(),color='r',linestyle=':')

scatter = ax1.scatter(
           x=over_50['age'],
           y=over_50['thalach'],
           c=over_50['target'])

ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(over_50['thalach'].mean(),color='b',linestyle='--');
```

```python
# Sharing the x-axis between two subplots 

# plot
fig, (ax,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                            sharex=True) # share the x axis

# add data
scatter = ax.scatter(
            x=over_50['age'],
            y=over_50['chol'],
            c= over_50['target'])

# customise the graph
ax.set(title='Heart Disease and Cholesterol Levels',
      
      ylabel='Cholesterol')

#adding legend
ax.legend(*scatter.legend_elements(),title='Target')

# adding the line 
ax.axhline(over_50['chol'].mean(),color='r',linestyle=':')

scatter = ax1.scatter(
           x=over_50['age'],
           y=over_50['thalach'],
           c=over_50['target'])

ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(over_50['thalach'].mean(),color='b',linestyle='--');

# adding title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold')
```
