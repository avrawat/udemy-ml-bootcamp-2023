### Two ways of plotting with matplotlib 

1. Pyplot - Less flexible
2. Object-oriented API - more flexible - widely used

https://matplotlib.org/stable/tutorials/introductory/lifecycle.html#sphx-glr-tutorials-introductory-lifecycle-pym

```python
## 1st method - Pyplot

fig = plt.figure() # creats a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method - Pyplot

fig = plt.figure() # create a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y) # add some data
plt.show()
```

```python
# 3rd method - recommend - OO API
fig, ax = plt.subplots()
ax.plot(x,y); #add some data
type(fig),type(ax)
```

## Matplotlib example workflow 

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # width * hight of the plot

# 3. Plot data - add data to axes
ax.plot(x,y)

# 4. Customise plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel="y-axis")

# 5. Save & show ( you can save the whole figure)
fig.savefig("exported-plots/sample-plot.png")
```

## Common types of Visualisation Plot 
### Making figures with NumPy arrays

### 1. Line Plot

```python
import numpy as np
```

```python
#create some data
# Return evenly spaced numbers over a specified interval.
x = np.linspace(1,10,100)
x[:10]
```

```python
#plot the data and create a line plot

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,x**2); 

ax.set(title="sample-plot2",xlabel="x-axis",ylabel='y-axis')
```

### 2. Scatter Plot

```python
# scatter plot

fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))

ax.set(title='Scatter Plot')
```

```python
# another scatter plot

fig,ax = plt.subplots()
ax.scatter(x,np.sin(x)) # sin wave
ax.plot();
```

### 3. Bar Plot

```python
# Using dictionary to create a plot

nut_butter_prices = {"Almond Butter":10,
                    "Peanut Butter":8,
                    "Cashew Butter":12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())

ax.set(title="Dan's Nut Butter Prices",ylabel="Price ($)");
```

```python
# horizontal bar graph

fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Butter Prices",xlabel="Prices ($)");
```

### 4. Histogram

```python
# make some data

# Return a sample (or samples) from the "standard normal" distribution.
x = np.random.randn(1000)
fig,ax = plt.subplots(1,2,figsize=(20,10))

ax[0].plot(x)
ax[1].hist(x)

ax[0].set(title="Normal Distribution",xlabel="Value",ylabel="Frequence");
ax[1].set(title="Normal Distribution",xlabel="Value",ylabel="Frequence");
```

### Two options for Subplot
#### Multiple plots at the same time



```python
# subplot option 1 - tuple way

fig, ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each diffrent axis

ax0.plot(x,x/2);
ax1.scatter(np.random.random(10),np.random.random(10))
ax2.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax3.hist(np.random.randn(1000))
```

```python
# subplot option 2 - list index

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each different way 
ax[0,0].plot(x,np.sin(x))
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Ploting from Pandas DataFrames

```python
import pandas as pd
```

```python
# make a dataframe
car_sales = pd.read_csv("./../data/car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),index=pd.date_range("05/05/2023",periods=1000))

# Cumulative Sum
ts = ts.cumsum()

ts.plot()
```
